```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# import seaborn as sns
```

```python
#insurance.csv'파일 읽어서 df에 할당하고, 상위 5개 항목 출력하기
df = pd.read_csv('data/insurance.csv')
df.head()
```

```python
#df의 요약(평균, 4분위수, 전체 갯수 등)정보 확인하기
df.describe()
```

```python
#신규 컬럼(double bmi)을 생성하고, bmi의 2배가 되는 값을 할당하기
df['double bmi'] = [i * 2 for i in df.loc[:,'bmi']]
df
```

```python
#신규 컬럼(debt)을 생성하고, 0으로 채우기
df['debt'] = 0
df
```

```python
#debt에 0부터 값이 증가하도록 채우기
df['debt'] = [i for i in range(len(df))]
df
```

```python
#1번 인덱스의 정보 출력
df.iloc[1]
```

```python
#'double bmi'열 삭제하기
df1 = df.drop('double bmi',axis=1)
df = df1
df
```

```python
#'age'열, 'sex'열만 출력하기

df.loc[:,['age','sex']]
```

```python
#나이를 1살씩 증가시키기
df['age'] = [i + 1 for i in df['age']]
df
```

```python
#charges의 표시형식을 소수점 아래 2자리로 바꾸기
df['charges']= [round(i,2) for i in df['charges']]
df
```

```python
#컬럼 정렬(알파벳 순서대로) : age, bmi, charges... 순으로 정렬
df.sort_index(axis=1)
```

```python
#값 정렬(age) : 나이 순으로 정렬
df.sort_values('age')
```

```python
#나이('age') 평균구하기
df['age'].mean()
```

```python
#보험료('charges')와 가장 상관관계가 높은 변수는?
```

```python
#남자, 여자가 각각 몇명있는지 구하기
male = df.value_counts('sex')['male']
female = df.value_counts('sex')['female']

print(f"남자 {male}명, 여자 {female}명")
```

```python
#자녀(children)가 3명인 데이터 출력하기
df[df['children'] == 3]
```

```python
#자녀(children)가 5명 이상인 데이터 출력하기
df[df['children']>=5]
```

```python
#bmi가 17 미만인 데이터 출력하기
df[df['bmi']<17]
```

```python
#성별에 따른 보험료 평균
male_ch_avg = df[df['sex'] == 'male']['charges'].mean()
female_ch_avg = df[df['sex'] == 'female']['charges'].mean()
print(f"남자 보험료 평균 : {male_ch_avg}")
print(f"여자 보험료 평균 : {female_ch_avg}")
```

```python
#흡연여부에 따른 보험료 평균
smoker_ch_avg = df[df['smoker'] == 'yes']['charges'].mean()
nosmoker_ch_avg = df[df['smoker'] == 'no']['charges'].mean()
print(f"흡연자 보험료 평균 : {smoker_ch_avg}")
print(f"비흡연자 보험료 평균 : {nosmoker_ch_avg}")
```

```python
#지역별 보험료 평균
southw_ch_avg = df[df['region'] == 'southwest']['charges'].mean()
northw_ch_avg = df[df['region'] == 'northwest']['charges'].mean()
southe_ch_avg = df[df['region'] == 'southeast']['charges'].mean()
northe_ch_avg = df[df['region'] == 'northeast']['charges'].mean()
print(f"남서 보험료 평균 : {southw_ch_avg}")
print(f"북서 보험료 평균 : {northw_ch_avg}")
print(f"남동 보험료 평균 : {southe_ch_avg}")
print(f"북동 보험료 평균 : {northe_ch_avg}")
```

```python
#흡연여부 + 지역별 데이터 평균
df.groupby(['smoker','region']).describe()
```

```python
#흡연여부 + 지역별 데이터 평균에서 'charges' 만 츨력하기
df.groupby(['smoker','region'])['charges'].describe()
```

```python
#흡연여부 + 지역별 데이터 크기(size)
df.groupby(['smoker','region']).size()
```

```python
#smoker 여부에 따라, group_val 컬럼에 '흡연자', '비흡연자' 표시하기
df['group_val'] = ['흡연자' if s == 'yes' else '비흡연자' for s in df['smoker']]
df
```
