```python
dates = ['2024-10-01', '2024-10-02', '2024-10-03', '2024-10-04', '2024-10-05']
prices = [100, 102, 101, 105, 103]
```

```python
plt.figure(figsize=(10,6))
plt.title('5일간 주식 가격')
plt.plot(dates,prices, marker='o', label='Stock price', color='blue')
plt.legend()
plt.xlabel('날짜')
plt.ylabel('가격')
plt.grid()
plt.ylim(95,110)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(dates,prices, marker='o', color='green')
plt.title('Stock Price - Subplot1')
plt.xticks(rotation = 45)
plt.subplot(1,2,2)
plt.bar(dates,prices, color='orange')
plt.title('Stock Price - Subplot2')
plt.xticks(rotation = 45)


plt.savefig('my_stock_price_chart.png')
```
