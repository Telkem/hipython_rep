# 새로운 데이터셋

Seaborn.tips 데이터 셋

```python
import seaborn as sns
tips_df = sns.load_dataset('tips')
tips_df.head()
```

```python
# 1변량 그래프 :히스토그램
fig = plt.figure(figsize=[12,5])
axes1 = fig.add_subplot(1,2,1)
axes2 = fig.add_subplot(1,2,2)
axes1.hist(data=tips_df, x='total_bill', bins=20)
axes2.hist(data=tips_df, x='tip', bins=20)
fig.tight_layout()
fig.show()
```

```python
#boxplot
fig = plt.figure(figsize=[7,5])
axes = fig.add_subplot(1,1,1)
axes.boxplot(data=tips_df, x='total_bill')
fig.show()
```

```python
fig = plt.figure(figsize=[7,5])
axes = fig.add_subplot(1,1,1)
axes.boxplot(x=[tips_df[tips_df['sex'] == 'Female']['total_bill'],tips_df[tips_df['sex'] == 'Male']['total_bill']],labels=["Female","Male"])
fig.show()
```

```python
fig = plt.figure(figsize=[7,5])
axes = fig.add_subplot(1,1,1)
axes.boxplot(x=[tips_df[tips_df['time'] == 'Lunch']['tip'],tips_df[tips_df['time'] == 'Dinner']['tip']],labels=["Lunch","Dinner"])
fig.show()
```

```python
fig = plt.figure(figsize=[10,5])
axes = fig.add_subplot(1,1,1)
axes.boxplot(x=
             [tips_df[tips_df['day'] == 'Thur']['total_bill'],
              tips_df[tips_df['day'] == 'Fri']['total_bill'],
              tips_df[tips_df['day'] == 'Sat']['total_bill'],
              tips_df[tips_df['day'] == 'Sun']['total_bill']],
             labels=["Thur","Fri", "Sat", "Sun"])
fig.show()
```

```python
fig = plt.figure(figsize=[10,5])
axes = fig.add_subplot(1,1,1)
axes.boxplot(x=
             [tips_df[(tips_df['day'] == 'Sat') & (tips_df['time'] == 'Lunch')]['tip'],
              tips_df[(tips_df['day'] == 'Sat') & (tips_df['time'] == 'Dinner')]['tip']],
             labels=["Lunch","Dinner"])
plt.title('토요일 점심 저녁 팁')
plt.show()
```

```python
# scatter, 산점도
fig = plt.figure()
axes = fig.add_subplot(1,1,1)
axes.scatter(
    data=tips_df,
    x='total_bill',
    y='tip'
)
plt.show()
```

```python
#다변량 표현
#성별에 따라 - 색상
#값의 크기에 따라 - 점의 크기
colors = {"Female":"red","Male":"blue"}
tips_df['tip_pct'] = tips_df['tip']/tips_df['total_bill']
fig = plt.figure()
axes = fig.add_subplot(1,1,1)
axes.scatter(
    data=tips_df,
    x='total_bill',
    y='tip',
    s=tips_df['tip_pct'] * 500,
    c=tips_df['sex'].map(colors),
    alpha=0.5
)
plt.show()
```

```python
#파이 차트 : 비율
#성별 비율 비교
sex_counts = tips_df['sex'].value_counts()
fig = plt.figure()
axes = fig.add_subplot(1,1,1)
axes.pie(
    sex_counts,
    colors=['blue','red'],
    labels=sex_counts.index,
    autopct='%1.1f%%',
    startangle=65
)
plt.show()
```

```python
sex_counts = tips_df['sex'].value_counts()
fig = plt.figure(figsize=[12,5])
axes1 = fig.add_subplot(1,2,1)
axes2 = fig.add_subplot(1,2,2)
axes1.pie(
    sex_counts,
    colors=['blue','red'],
    labels=sex_counts.index,
    autopct='%1.1f%%',
    startangle=65
)

axes2.scatter(
    data=tips_df,
    x='total_bill',
    y='tip',
    s=tips_df['tip_pct'] * 500,
    c=tips_df['sex'].map(colors),
    alpha=0.5
)
plt.show()
```
